```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#Preprocessing techniques will vary for different datasets

pos = pd.read_csv('freeletics/pos1.csv')
neg = pd.read_csv('freeletics/neg1.csv')

pos_text = pos.values[:,[2]]
neg_text = neg.values[:,[2]]

y_pos = np.ones((len(pos_text),1))
y_neg = np.zeros((len(neg_text),1))

x_data = np.vstack((pos_text,neg_text))
y_data = np.vstack((y_pos,y_neg))
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.25)
train = np.hstack((X_train,y_train))
test = np.hstack((X_test,y_test))
```

```python
train_df = pd.DataFrame(train)
test_df = pd.DataFrame(test)

train_df.columns = ['text','label']
test_df.columns = ['text','label']


vectorizer = TfidfVectorizer(max_features=10000)
train_vec = vectorizer.fit_transform(train_df.text.values)
test_vec = vectorizer.transform(test_df.text.values)

#print(vectorizer.get_feature_names())
```
